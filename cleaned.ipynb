```python
# Fungsi untuk menghilangkan outlier berdasarkan mean dan standar deviasi
def remove_outliers_by_deviation(data, column):
    mean = data[column].mean()
    std_deviation = data[column].std()
    return data[abs(data[column] - mean) <= 3 * std_deviation]
```

```python
# Menampilkan outlier sebelum dan sesudah penghapusan
def show_outliers_by_deviation(data, column):
    outliers_before = data[~data.index.isin(remove_outliers_by_deviation(data, column).index)]
    print(f"Outlier dalam kolom '{column}' sebelum penghapusan:")
    print(outliers_before[[column, 'Sales', 'Quantity', 'Discount', 'Profit']])
    print("\nOutlier dalam kolom '{}' setelah penghapusan:".format(column))
    print(data[~data.index.isin(outliers_before.index)][[column, 'Sales', 'Quantity', 'Discount', 'Profit']])
```

```python
# Menampilkan outlier dalam bentuk tabel
for column in ['Sales', 'Quantity', 'Discount', 'Profit']:
    print(f"Outlier dalam kolom '{column}' sebelum penghapusan:")
    display(df[~df.index.isin(remove_outliers_by_deviation(df, column).index)][[column, 'Sales', 'Quantity', 'Discount', 'Profit']])
```

```python
# Menghapus outlier
for column in ['Sales', 'Quantity', 'Discount', 'Profit']:
    df = remove_outliers_by_deviation(df, column)
```
